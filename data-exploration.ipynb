```python
from statistics import correlation

import kagglehub
import pandas as pd
from pathlib import Path

from matplotlib import pyplot as plt

dataset_path = kagglehub.dataset_download(
    "dylanjcastillo/7k-books-with-metadata"
)

print(dataset_path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{dataset_path}/books.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
 import numpy as np

 books["missing_description"] = np.where(books["description"].isna(), 1, 0)
 books["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_of_interest = [
    "num_pages",
    "age_of_book",
    "missing_description",
    "average_rating"
]

correlation_matrix = books[columns_of_interest].corr(method="pearson")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))

heatmap = sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar_kws={"label": "Pearson Correlation"}
)

heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
books_missing = books[
    books["description"].notna() &
    books["num_pages"].notna() &
    books["average_rating"].notna() &
    books["published_year"].notna()
]
```

```python
 books_missing["categories"].value_counts() .reset_index(name="count").sort_values(by="count",ascending=False)
```

```python
books_missing["words_in_description"] = books_missing["description"].str.split().str.len()
```

```python
books_missing
```

```python
books_missing.loc[books_missing["words_in_description"].between(25,34), "description"]
```

```python
books_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

```python
books_missing_25_words
```

```python
books_missing_25_words["title_and_subtitle"] = np.where(
    books_missing_25_words["subtitle"].isna(),
    books_missing_25_words["title"],
    books_missing_25_words["title"].astype(str) + ": " +
    books_missing_25_words["subtitle"].astype(str)
)
```

```python
books_missing_25_words
```

```python
books_missing_25_words["tagged_description"] =  books_missing_25_words["isbn13"].astype(str) + books_missing_25_words["description"].astype(str)
```

```python
books_missing_25_words
```

```python
books_cleaned = books_missing_25_words.drop(
    ["subtitle", "words_in_description", "missing_description", "age_of_book"],
    axis=1
)

books_cleaned.to_csv("books_cleaned.csv", index=False)
```
